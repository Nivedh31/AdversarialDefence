```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gaussian Blur
input_image_path = r'attack_test\adv_images\advairplane.jpg'
image = cv2.imread(input_image_path)

gaussian_blur = cv2.GaussianBlur(image, (5, 5), 0)
```

```python
# Median Filtering
median_filtered = cv2.medianBlur(gaussian_blur, 5)
```

```python
plt.figure(figsize=(12, 4))
    
# Original Image
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
    
# Gaussian Blurred Image
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(gaussian_blur, cv2.COLOR_BGR2RGB))
plt.title('Gaussian Blurred')
plt.axis('off')
    
# Median Filtered Image
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(median_filtered, cv2.COLOR_BGR2RGB))
plt.title('Median Filtered')
plt.axis('off')
    
plt.tight_layout()
plt.show()

# Save the final processed image
# output_image_path = r'attack_test\adv_denoise\gaus_mdnfl_img\gaus_mdnfl_truck.jpg'
# cv2.imwrite(output_image_path, median_filtered)
```
