```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure
from PIL import Image, ImageFilter
```

```python
def apply_adaptive_contrast_enhancement(image):
    # Convert the image to numpy array
    image_array = np.array(image)

    # Apply adaptive histogram equalization
    enhanced_image_array = exposure.equalize_adapthist(image_array, clip_limit=0.03)

    # Convert the enhanced numpy array back to PIL Image
    enhanced_image = Image.fromarray((enhanced_image_array * 255).astype(np.uint8))

    return enhanced_image
```

```python
# Load the image
image_path = r'attack_test\adv_images\advbird.jpg'
image = Image.open(image_path)

# Apply adaptive contrast enhancement
enhanced_image = apply_adaptive_contrast_enhancement(image)

# Save the enhanced image
#enhanced_image_path = r'attack_test\org_denoise\ace_img\ace_truck.jpg'
#enhanced_image.save(enhanced_image_path)
```

```python
# Display the original and enhanced images
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(enhanced_image)
plt.title('Enhanced Image')

plt.show()
```
