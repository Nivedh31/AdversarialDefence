```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

# Load the image
image = Image.open(r'attack_test\adv_images\advdeer.jpg')

# Apply Gaussian blur
blurred_image = image.filter(ImageFilter.GaussianBlur(radius=1))

# Display the original and preprocessed images side by side
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Preprocessed Image')
plt.imshow(blurred_image)
plt.axis('off')

plt.show()

#processed_image_path = r'attack_test\adv_denoise\gaus_img\procairplane.jpg'
#blurred_image.save(processed_image_path)
```
