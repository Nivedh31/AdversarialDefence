```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Input and output image paths
input_image_path = r'attack_test\adv_images\advdog.jpg'

# Load the input image
input_image = cv2.imread(input_image_path)

# Kernel size for median filtering
kernel_size = 3

# Apply median filtering
filtered_image = cv2.medianBlur(input_image, kernel_size)
    
# Display the input and output images using matplotlib
plt.figure(figsize=(12, 6))

# Input Image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
plt.title('Input Image')
plt.axis('off')



# Output Image (Filtered)
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(filtered_image, cv2.COLOR_BGR2RGB))
plt.title('Filtered Image')
plt.axis('off')

plt.show()

# Save the filtered image
# output_image_path = r'attack_test\org_denoise\mdnfl_img\mdfltruck.jpg'
# cv2.imwrite(output_image_path, filtered_image)
```
